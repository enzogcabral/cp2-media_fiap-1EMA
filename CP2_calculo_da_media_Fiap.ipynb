**Calcular a media dos checkpoints da fiap**



```python
# calculo de media dos checkpoints
# 3 CPs obrigatorias ----> input
CP1, CP2, CP3 = input("insira as notas dos checkpoints: ").split()
CP1 = float(CP1)
CP2 = float(CP2)
CP3 = float(CP3)
```

```python
# eliminar menor nota
D1 = min(CP1, CP2, CP3)
D1
```

```python
# calcular media das duas maiores notas
CPM1 = (CP3 + CP2 + CP1 - D1) / 2
CPM1
```

Calculo da media da Sprint

```python
SP1, SP2 = input("insira as notas da sprint: ").split()
SP1 = float(SP1)
SP2 = float(SP2)
SPF1 = (SP1 + SP2) / 2
SPF1
```

Valor da Global Solution

```python
GS1 = input("insira a nota da GS: ")
GS1 = float(GS1)
GS1
```

Calculo Da media do 1 semestre

```python
MS1 = GS1*0.6 + SPF1*0.2 + CPM1*0.2
MS1
```
